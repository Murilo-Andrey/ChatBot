```python
from google.colab import userdata
chat_api = userdata.get('AtividadeAula')
```

```python
from logging import debug
import gradio as gr # Added import for gr
from groq import Groq # Added import for Groq if not already present

client = Groq(
    api_key=chat_api
)

def assistente_quimicoefisico(user_prompt:str) -> str:

  try:
    if user_prompt.strip()=="15":
      return "Encerrando ChatBot"

    completion = client.chat.completions.create(
      model="llama-3.1-8b-instant",
      messages=[
          {
              "role": "system","content": "Você é um assistente especialista em química e física."
          },
          {
              "role": "user","content": user_prompt
          }
      ],
      temperature = 0.5,
      max_tokens = 1024,
      top_p = 1,
      stream = False
    )

    return completion.choices[0].message.content
  except Exception as e:
    print("\n Erro no Assistente em Química e Física")
    return f"Erro: {e}"

# Move iface definition outside the function and try-except block to be globally accessible
iface = gr.Interface(
    fn = assistente_quimicoefisico,
    inputs = gr.Textbox(lines=10, placeholder="Digite sua Pergunta sobre Física ou Química"),
    outputs = gr.Textbox(lines=10, label="Resposta"),
    title = "Assistente de Química e Física",
    description = "Digite sua pergunta sobre química ou física",
    live = False
)
```

```python
iface.launch(debug=True, show_error=True)
```
